```python
#input: playlist id
#gets json of this playlist, extracts the song ids for each song in playlist. 
#for each song, gets track information and audio features and converts them into a dataframe and csv
def playlist_to_dataset(playlist_id):
    playlist = sp.playlist(playlist_id)
    ids=[]

    for item in playlist['tracks']['items']:
            s_id = item['track']['id']
            ids.append(s_id)
            
    track_info = {'song_id':[], 'name':[], 'album':[], 'artist':[], 'release_date':[],
            'length':[], 'popularity':[], 'genre':[], 'danceability':[], 'acousticness':[],
            'danceability':[], 'energy':[], 'instrumentalness':[], 'liveness':[],
            'loudness':[], 'speechiness':[], 'tempo':[], 'time_signature':[]}
    for id in ids:
        metadata = sp.track(id)
        features = sp.audio_features(id)

        # metadata
        track_info['song_id'].append(id)
        track_info['name'].append(metadata['name'])
        track_info['album'].append(metadata['album']['name'])
        track_info['artist'].append(metadata['album']['artists'][0]['name'])
        track_info['release_date'].append(metadata['album']['release_date'])
        track_info['length'].append(metadata['duration_ms'])
        track_info['popularity'].append(metadata['popularity'])
        track_info['genre'].append(metadata['genre'])
  
        # audio features
        track_info['acousticness'].append(features[0]['acousticness'])
        track_info['danceability'].append(features[0]['danceability'])
        track_info['energy'].append(features[0]['energy'])
        track_info['instrumentalness'].append(features[0]['instrumentalness'])
        track_info['liveness'].append(features[0]['liveness'])
        track_info['loudness'].append(features[0]['loudness'])
        track_info['speechiness'].append(features[0]['speechiness'])
        track_info['tempo'].append(features[0]['tempo'])
        track_info['time_signature'].append(features[0]['time_signature'])


    dataset = pd.DataFrame.from_dict(track_info)
    return dataset

#the countries and their respective playlist ids for their official Top 50 Spotify playlist
top_50={'USA': '37i9dQZEVXbLRQDuF5jeBp', 'Japan': '37i9dQZEVXbKXQ4mDTEBXq', 'Mexico': '37i9dQZEVXbO3qyFxbkOE1', 
        'Spain': '37i9dQZEVXbNFJfN1Vw8d9', 'France': '37i9dQZEVXbIPWwFssbupI', 'Argentina': '37i9dQZEVXbMMy2roB9myp',
        'Philippines': '37i9dQZEVXbNBz9cRCSFkY', 'South Korea':'37i9dQZEVXbNxXF4SkHj9F',
       'Brazil': '37i9dQZEVXbMXbN3EUUhlg'}

#creates csv for each coutry's top 50 playlist
for country in top_50:
    playlist_to_dataset(top_50[country]).to_csv(country + '_Top_50.csv')
print('----------------------------------')
```

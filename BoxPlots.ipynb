## Box Plot Analysis of Country's Top 50 song data


```python
import json
import spotipy
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sb
import requests
from spotipy.oauth2 import SpotifyClientCredentials

#stores the dataframes for each coutry
dataframes = {'USA': [], 'Japan': [], 'Mexico': [], 
        'Spain': [], 'France': [], 'Argentina': [],
        'Philippines': [], 'South Korea':[], 'Brazil': []}

#reads each country's csv into the dictionary
for country in dataframes:
    dataframes[country] = pd.read_csv(country + '_Top_50.csv')
```

```python
#audio analysis attributes
box_params = ['length', 'popularity', 'danceability', 'acousticness',
            'danceability', 'energy', 'instrumentalness', 'liveness',
            'loudness', 'speechiness', 'tempo']

#for each audio attribute, makes a boxplot and prints its mean, standard deviation, minimum, 25% percentile, 50% percentile, 75% percentile , maximum for each country
for param in box_params:
    #x axis
    param_list=[]
    #y axis
    countries=[]
    
    #for each country, adds the values of the audio attribute and removes the zero values
    for country in dataframes:
        df = dataframes[country]
        df = df[df[param] != 0]
        print(country)
        print(df[param].describe())
        param_list.extend(df[param].values.tolist())
        countries.extend(np.full(shape=len(df[param]), fill_value=country).tolist())

    print(param + " total mean" + np.mean(param_list))
    plt.figure(figsize=(10,6))
    ax = sb.boxplot(x=param_list, y=countries)
    plt.xlabel(param.capitalize())
    plt.ylabel('')
    plt.title('Song '+ param.capitalize() + ' by Country', fontweight='bold', fontsize=18)
    plt.show()
```
